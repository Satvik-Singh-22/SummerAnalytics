### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
median_mpg= data['mpg'].median()
above_median= data[data['mpg'] > median_mpg]
max_hp_per_wt= above_median['hp_per_weight'].max()
best_cars= above_median[above_median['hp_per_weight'] == max_hp_per_wt]
best_cars
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
avg_mpg_per_year= data.groupby(['model_year','origin'])['mpg'].mean().unstack()
plt.figure(figsize=(10, 6))
for origin in avg_mpg_per_year.columns:
    plt.plot(avg_mpg_per_year.index, avg_mpg_per_year[origin], label=origin)

# Labeling
plt.title("Evolution of Average MPG over Years by Origin")
plt.xlabel("Model Year")
plt.ylabel("Average MPG")
plt.legend(title='Origin')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
data['horsepower'] = pd.to_numeric(data['horsepower'], errors='coerce')
data['weight'] = pd.to_numeric(data['weight'], errors='coerce')
data['mpg'] = pd.to_numeric(data['mpg'], errors='coerce')
hue_order = ['japan', 'europe', 'usa']
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=data,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    hue_order=hue_order,
    sizes=(20, 200),
    alpha=0.7
)
plt.title("Horsepower vs. Weight of Cars Colored by Origin and Sized by MPG", fontsize=14)
plt.xlabel("Horsepower")
plt.ylabel("Weight")
plt.legend(title="Origin", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
